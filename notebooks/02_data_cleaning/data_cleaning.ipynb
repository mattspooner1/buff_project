```python
import pandas as pd
import numpy as np
import requests
from functools import reduce
```

```python
def dataclean(patch):

    df1 = pd.read_csv ('11.'+str(patch)+'irontogold.csv')
    df2 = pd.read_csv ('11.'+str(patch)+'plattogm.csv')
    df3 = pd.read_csv ('11.'+str(patch)+'challenger.csv')
    changes = pd.read_csv('11.'+str(patch+1)+'changes.csv')
    
    
    irontogold = pd.merge(left=df1, right=changes, left_on='champ', right_on='champ')
    plattogm = pd.merge(left=df2, right=changes, left_on='champ', right_on='champ')
    challenger = pd.merge(left=df3, right=changes, left_on='champ', right_on='champ')
    
    
    irontogold.to_csv('11.'+str(patch)+'irontogold.csv',index=False)
    plattogm.to_csv('11.'+str(patch)+'plattogm.csv',index=False)
    challenger.to_csv('11.'+str(patch)+'challenger.csv',index=False)
    
```

```python
def getcleancsv ():
    for X in range (1,19):
        dataclean(X)
```
