```python
import pandas as pd
import numpy as np
import requests
```

```python
def dataclean(patch):

    df1 = pd.read_csv (str(patch)+'iron,bronze,silver,gold.csv')
    df2 = pd.read_csv (str(patch)+'platinum,diamond,master,grandmaster.csv')
    df3 = pd.read_csv (str(patch)+'challenger.csv')
    changes = pd.read_csv(str(patch)+'changes.csv')
    
    ironbronzesilvergold = pd.merge(left=df1, right=changes, left_on='champ', right_on='champ')
    platinumdiamondmastergrandmaster = pd.merge(left=df2, right=changes, left_on='champ', right_on='champ')
    challenger = pd.merge(left=df3, right=changes, left_on='champ', right_on='champ')
    
    ironbronzesilvergold = ironbronzesilvergold.drop(['Unnamed: 0_x'],axis =1)
    ironbronzesilvergold = ironbronzesilvergold.drop(['Unnamed: 0_y'],axis =1)
    
    platinumdiamondmastergrandmaster = platinumdiamondmastergrandmaster.drop(['Unnamed: 0_x'],axis =1)
    platinumdiamondmastergrandmaster = platinumdiamondmastergrandmaster.drop(['Unnamed: 0_y'],axis =1)
    
    challenger = challenger.drop(['Unnamed: 0_x'],axis =1)
    challenger = challenger.drop(['Unnamed: 0_y'],axis =1)
    
    ironbronzesilvergold.to_csv(str(patch)+'iron,bronze,silver,gold.csv',index=False)
    platinumdiamondmastergrandmaster.to_csv(str(patch)+'platinum,diamond,master,grandmaster.csv',index=False)
    challenger.to_csv(str(patch)+'challenger.csv',index=False)
    
```

```python
def getcleancsv ():
    for X in range (2,17):
        dataclean("11."+str(X))
```
